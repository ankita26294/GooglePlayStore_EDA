```python
df_copy=df.copy()
# make copy of df it denoted as df_copy
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
df_copy.head()
# drop string value row
```

```python
df_copy['Reviews']=df_copy['Reviews'].astype('int')
# convert Reviews datatype object to integer
```

```python
for i in df_copy['Size'].str:
    print(i)
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
# replace all M into '000'
```

```python
df_copy['Size']
```

```python
df_copy['Size']=df_copy['Size'].str.replace('k','')
# replace all k to ''
```

```python
df_copy['Size'].unique()
# removing M,k
```

```python
# replace Varies with device by nan but here all numbers in string thats why here write str(np.nan)
df_copy['Size']=df_copy['Size'].str.replace('Varies with device',str(np.nan))
df_copy['Size']
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'].dtype
```

```python
df_copy["Size"]=df_copy["Size"].astype('float')
df_copy["Size"]
# converting object to float byusing astype
```

```python
df_copy['Size'].head()
```

```python
df_copy['Size'][2]*1000
```

```python
for i in df_copy['Size']:
    if i <10:
        df_copy['Size']=df_copy['Size'].replace(i,i*1000)
```

```python
df_copy['Size']=df_copy['Size']/1000
df_copy['Size']
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for col in cols_to_clean:
        df_copy[col]=df_copy[col].str.replace(item,'')
```

```python
df_copy['Price'].unique()
```

```python
# change data type of Installs:object to int
df_copy['Installs']=df_copy['Installs'].astype('int')
```

```python
# change data type of Price:object to float
df_copy['Price']=df_copy['Price'].astype('float')
```

```python
df_copy['Last updated']=pd.to_datetime(df_copy['Last Updated'])
# convert into datetime.eg.January 7, 2018 to 1/7/2018
```

```python
df_copy['Last updated']
```

```python
# create Day column
df_copy['Day']=df_copy['Last updated'].dt.day
```

```python
# create month column
df_copy['month']=df_copy['Last updated'].dt.month
```

```python
# create year column
df_copy['Year']=df_copy['Last updated'].dt.year
```
